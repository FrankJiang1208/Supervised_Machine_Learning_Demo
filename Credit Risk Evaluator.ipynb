```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
pd.set_option('display.max_columns', None)
train_df
```

```python
# Convert categorical data to numeric and separate target feature for training data
training_y=train_df["loan_status"]
training_X=train_df.drop(columns="loan_status",axis=1)
training_X=pd.get_dummies(training_X)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_y=test_df["loan_status"]
testing_X=test_df.drop(columns="loan_status",axis=1)
testing_X=pd.get_dummies(testing_X)
```

```python
training_X
```

```python
# add missing dummy variables to testing set
testing_X["debt_settlement_flag_Y"]=0
```

I hypothesize that random forest classifier would work better since linear regression model is used to predict simpler scenario with less independent variables whereas random forest classifier will not be heavily influenced by quantity of variables.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(training_X, training_y)
print(f"Training Data Score: {classifier.score(training_X, training_y)}")
print(f"Testing Data Score: {classifier.score(testing_X, test_y)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
clf = RandomForestClassifier(random_state=1, n_estimators=50).fit(training_X, training_y)
print(f'Training Score: {clf.score(training_X, training_y)}')
print(f'Testing Score: {clf.score(testing_X, test_y)}')
```

My prediction is that scaling will improve both Classifications at least by 10 percent since the weight of the independent variables are very different.

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(training_X)
X_train_scaled = scaler.transform(training_X)
X_test_scaled = scaler.transform(testing_X)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression()
classifier.fit(X_train_scaled, training_y)
print(f"Training Data Score: {classifier.score(X_train_scaled, training_y)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, test_y)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier( n_estimators=50).fit(X_train_scaled, training_y)
print(f'Training Score: {clf.score(X_train_scaled, training_y)}')
print(f'Testing Score: {clf.score(X_test_scaled, test_y)}')
```

My prediction is inaccurate for the testing phase, while it is correct for the training phase. I still think Forest Classifier will work better once the parameters are tweaked.
